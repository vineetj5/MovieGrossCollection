```python

# Find and print the top-level keys
print("Top-level keys:", scriptJson94.keys())

# If you want to find the keys at all levels (nested), you can define a recursive function
def get_all_keys(d):
    keys = []
    for key, value in d.items():
        keys.append(key)
        if isinstance(value, dict):
            keys.extend(get_all_keys(value))  # Recursively get keys from nested dictionaries
    return keys

# Find all keys (including nested ones)
all_keys = get_all_keys(scriptJson94)
print("All keys:", all_keys)
```

```python
def find_elements_by_key(d, target_key):
    elements = []

    # If the current object is a dictionary
    if isinstance(d, dict):
        for key, value in d.items():
            # If the current key matches the target key
            if key == target_key:
                elements.append(value)
            # Recursively search for the target key in nested dictionaries
            elements.extend(find_elements_by_key(value, target_key))
    
    # If the current object is a list, search in each element
    elif isinstance(d, list):
        for item in d:
            elements.extend(find_elements_by_key(item, target_key))

    return elements

# Find all elements under the key 'zipcode'
edges = find_elements_by_key(scriptJson94, 'edges')

print(edges)
```

```python
edge = edges[0]

dictedge = edge[10]

movieMetaData = dictedge['node']['description']['originalText']['plaidHtml']

pattern = r"([A-Za-z\s]+)\s*:\s*([\d.]+cr|[A-Za-z\s]+)"

# Find all matches using regex
matches = re.findall(pattern, movieMetaData)

# Print the matched fields and their values
for field, value in matches:
    print(f"{field.strip()}: {value.strip()}")
```

```python
edge = edges[0]
pattern = r"([A-Za-z\s]+)\s*:\s*([\d.]+cr|[A-Za-z\s]+)"
movieMetaList=[]
for iterEdge in edge:
    movieMetaData = iterEdge['node']['description']['originalText']['plaidHtml']
    matches = re.findall(pattern, movieMetaData)
    movieMetaDict={}
    for field, value in matches:
        movieMetaDict[field.strip()]=value.strip()
    movieMetaList.append(movieMetaDict)
movieMetaList
```
